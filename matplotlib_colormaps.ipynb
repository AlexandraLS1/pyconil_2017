```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
cmaps = plt.colormaps()
cmaps_forward = [ n for n in plt.colormaps() if not( n.endswith('_r')) ]
for cmap in cmaps_forward:
    cm = plt.get_cmap(cmap)
    if 'colors' in dir( cm ):
        colors = getattr( cm, 'colors' ) 

        if not colors: continue
        print( "{m} [{n}]".format( m = cmap, n = len( colors ) ) )

        cols = 3 if len( colors ) % 3 == 0 else 4
        if len( colors ) % 5 == 0: cols = 5
        if len( colors ) == 256: cols = 16
        rows  = math.ceil( len( colors ) / cols )
        cells = cols * rows

        img = np.zeros( (rows * cols, 3) )
        img[:len(colors)] = colors

        img = img.reshape( ( rows, cols, 3 ) )

        figsize = (12,7) if rows > cols else (7,12)
        if cols == 3 and rows == 3: figsize = (6, 6)

        f, ax = plt.subplots( figsize = figsize )
        ax.grid('off')
        ax.axis('off')
        ax.imshow( img )

    else:
        f, ax = plt.subplots( figsize = (8,8) )
        ax.axis('off')
        img = np.linspace(0,1,256).reshape((16,16))
        print( "{m} [{n}]".format( m = cmap, n = 256 ) )
        plt.imshow( img, cmap = plt.get_cmap( cmap ) )
        
    plt.show()
```
